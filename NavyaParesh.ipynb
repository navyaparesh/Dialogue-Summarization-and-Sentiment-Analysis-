## IMPORTING NECESSARY LIBRARIES AND DATA

```python
from googletrans import Translator
import re
from transformers import pipeline
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
nltk.download('vader_lexicon')
```

```python
# Hindi dialog
hindi_dialog = """
RA: नमस्ते कुमार, मैं एएस वाई जेड फाइनेंस से बोल रहा हूं। आपके लोन के बारे में बात करनी थी।
B: हां, बोलिए। यह क्या बात है?
RA: सर, आपका पिछले महीने का EMI अभी तक नहीं आया है। कोई समस्या है?
B: हां, थोड़ी दिक्कत है। मेरी नौकरी चली गई है और मैं नया काम ढूंढ रहा हूं।
RA: ओह, यह तो बुरा हुआ। लेकिन सर, आपको समझना होगा कि लोन का भुगतान समय पर करना बहुत जरूरी है।
B: मैं समझता हूं, लेकिन अभी मेरे पास पैसे नहीं हैं। क्या कुछ समय मिल सकता है?
RA: हम समझते हैं आपकी स्थिति। क्या आप अगले हफ्ते तक कुछ भुगतान कर सकते हैं?
B: मैं कोशिश करूंगा, लेकotsa EMI नहीं दे पाऊंगा। क्या आधा भुगतान चलेगा?
RA: ठीक है, आधा भुगतान अगले हफ्ते तक कर दीजिए। बाकी का योजना है आपका?
B: मुझे उम्मीद है कि अगले महीने तक मुझे नया काम मिल जाएगा। तब मैं बाकी बकाया चुकाऊंगा।
RA: ठीक है। तो हम ऐसा करते हैं - आप अगले हफ्ते तक आधा EMI जमा कर दीजिए, और अगले महीने के 15 तारीख तक बाकी का भुगतान कर दीजिए। यह आपको स्वीकार है?
B: हां, यह ठीक रहेगा। मैं इस योजना का पालन करने की पूरी कोशिश करूंगा।
RA: बहुत अच्छा। मैं आपको एक SMS भेज रहा हूं जिसमें भुगतान की डेट्स होंगी। कृपया इसका पालन करें और समय पर भुगतान करें।
B: ठीक है, धन्यवाद आपके समझने के लिए।
RA: आपका स्वागत है। अगर कोई और सवाल हो तो मुझे बताइएगा। अलविदा।
B: अलविदा।
"""
```

## Translate Hindi to English

```python
def translate_to_english(hindi_text):
    translator = Translator()
    english_text = translator.translate(hindi_text, dest='en').text
    return english_text
```

```python
def format_translated_text(english_text):
    lines = english_text.split('\n')
    formatted_lines = []

    for line in lines:
        match = re.match(r'^(.*?): (.*)$', line)
        if match:
            name, speech = match.groups()
            formatted_lines.append(f"{name}: {speech}")
        else:
            formatted_lines.append(line)

    return '\n'.join(formatted_lines)
```

```python
# Translate to English
english_dialog = translate_to_english(hindi_dialog)

# Format the translated text
formatted_english_dialog = format_translated_text(english_dialog)
```

```python
print("English Translation:")
print(formatted_english_dialog)
```

## DIALOUGE SUMMARIZATION USING BART MODEL

```python
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")
```

```python
# Summarize the dialogue
summary = summarizer(formatted_english_dialog,  max_length=100, min_length=10, do_sample=False)
print("\nSummary:")
print(summary[0]['summary_text'])
```

## SENTIMENT ANALYSIS

```python
# Initialize VADER sentiment analyzer
sia = SentimentIntensityAnalyzer()
```

```python
# Calculating Sentiment
def calculate_sentiment(text):
    sentiment_scores = sia.polarity_scores(text)
    return sentiment_scores['compound']
```

```python
# Analyze sentiment
def analyze_sentiment(dialogue):
    lines = dialogue.split('\n')

    recovery_agent_sentiment = []
    borrower_sentiment = []

    current_role = None

    for line in lines:
        match = re.match(r'^(.*?): (.*)$', line)
        if match:
            name, speech = match.groups()

            if name == 'RA':
                current_role = 'RA'
                recovery_agent_sentiment.append(calculate_sentiment(speech))
            elif name == 'B':
                current_role = 'B'
                borrower_sentiment.append(calculate_sentiment(speech))
            else:
                recovery_agent_sentiment.append(calculate_sentiment(line.strip()))

        # Add neutral score if there's no role specified
        if not current_role:
            recovery_agent_sentiment.append(0)
            borrower_sentiment.append(0)

    return recovery_agent_sentiment, borrower_sentiment

ra_sentiment, b_sentiment = analyze_sentiment(formatted_english_dialog)

print("\nRecovery Agent Sentiment Analysis:")
for i, score in enumerate(ra_sentiment):
    print(f"Line {i+1}: {score}")

print("\nBorrower Sentiment Analysis:")
for i, score in enumerate(b_sentiment):
    print(f"Line {i+1}: {score}")
```

```python
# Calculate average sentiments of the agent and borrower
avg_ra_sentiment = sum(ra_sentiment) / len(ra_sentiment)
avg_b_sentiment = sum(b_sentiment) / len(b_sentiment)

print(f"\nAverage Recovery Agent Sentiment: {avg_ra_sentiment}")
print(f"Average Borrower Sentiment: {avg_b_sentiment}")

if avg_ra_sentiment > 0:
    print("Overall Recovery Agent Sentiment: Positive")
elif avg_ra_sentiment < 0:
    print("Overall Recovery Agent Sentiment: Negative")
else:
    print("Overall Recovery Agent Sentiment: Neutral")

if avg_b_sentiment > 0:
    print("Overall Borrower Sentiment: Positive")
elif avg_b_sentiment < 0:
    print("Overall Borrower Sentiment: Negative")
else:
    print("Overall Borrower Sentiment: Neutral")
```
